```python
import pandas as pd
import os
from sklearn.model_selection import train_test_split
```

```python
image_path_list = []
label_list = []


for image_object in os.listdir(f"images/uais_logo_images/"):
    image_path_list.append(f"images/uais_logo_images/{image_object}")
    label_list.append(f"yes")

for image_object in os.listdir(f"images/Random images/"):
    image_path_list.append(f"images/Ramdom images/{image_object}")
    label_list.append(f"no")


df = pd.DataFrame()


df["image_path"] = image_path_list
df["label"] = label_list


# now split this main data to train and test
# Define the split ratio
test_ratio = 0.20 # 20% of data will go to test

# split the data
train_df, test_df = train_test_split(df, test_size=test_ratio, 
                                          stratify=df['label'], 
                                          random_state=42)


print(f"Original dataset shape: {df.shape}")
print(f"Train dataset shape: {train_df.shape}")
print(f"Test dataset shape: {test_df.shape}")

```
